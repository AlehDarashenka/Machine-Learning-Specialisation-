```python
import pandas as pd
df = pd.read_csv('checkins.dat', sep='|', skipinitialspace = True)\

```

```python
df.columns
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
df.info()


```

```python
df.drop([u'id    ', u'user_id ', u'venue_id ', u'created_at      '], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.cluster import MeanShift
```

```python
model = MeanShift(bandwidth=0.1)
model.fit(df[:100001])
```

```python
len(model.cluster_centers_)
```

```python
model.labels_
```

```python
label = model.predict(df)
```

```python
from collections import Counter
cnt = Counter()
for i in label:
    cnt[i]+=1

    
```

```python
dict_cnt = dict(cnt)
```

```python
clusters = [key for key in cnt if cnt[key]> 15]
```

```python
print len(clusters)
```

```python
coordinates = {data : model.predict(data) if model.predict(data) in clusters}
```

```python
center_clusters = [model.cluster_centers_[label] for label in clusters]
```

```python
len(center_clusters)
```
